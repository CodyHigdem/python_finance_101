```python
import numpy as np
gdp_cap = [100, 200, 141, 43]
life_exp = [65, 82, 73, 44]
population = [23, 45, 210, 1300]
```

```python
plt.scatter(gdp_cap, life_exp, s=population)
```

```python
double_pop = np.array(population) * 4
```

```python
plt.scatter(gdp_cap, life_exp, s=double_pop)
```

```python
country_colors = {
    'Oceania': 'black',
    'Europe': 'red',
    'Americas': 'yellow',
    'Asia': 'green'
}
```

```python
plt.scatter(x=gdp_cap, y=life_exp, s=double_pop, c=country_colors)
```
